```python
import pandas as pd
import os
import matplotlib.pyplot as plt
from supervenn import supervenn
datapath = "../data"
# load labeling
df_help = pd.read_csv(os.path.join(datapath, "Kidney_HELP.csv"), index_col=0)
df_ogee = pd.read_csv(os.path.join(datapath, "Kidney_OGEE.csv"), index_col=0)
df_fiper = pd.read_csv(os.path.join(datapath, "Kidney_FiPer.csv"), index_col=0)
df_adam = pd.read_csv(os.path.join(datapath, "Kidney_AdAM.csv"), index_col=0)
# load benchmarks

color_cycle = plt.rcParams['axes.prop_cycle'].by_key()['color']
color_cycle[0] = '#009ACD'
sets = [
        set(df_help[df_help['label']=='E'].index.values),
        set(df_fiper[df_fiper['label']=='E'].index.values),
        set(df_adam[df_adam['label']=='E'].index.values),
        set(df_ogee[df_ogee['label']=='E'].index.values),
       ]
labels = ['HELP', 'FiPer', 'ADaM', 'OGEE', 'HELP']
plt.figure(figsize=(20, 5))
color_cycle = plt.rcParams['axes.prop_cycle'].by_key()['color']
color_cycle[0] = '#009ACD'
supervenn(sets, labels, widths_minmax_ratio=0.05, side_plots='right', color_cycle=color_cycle)
plt.ylabel('Kidney-specific EGs', fontsize = 10)
#plt.savefig("labelling_svenn.jpg", dpi=600)
```
