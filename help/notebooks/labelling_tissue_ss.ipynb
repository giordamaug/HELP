# Load the CRISPR data file

```python
from help.utility.selection import filter_crispr_by_model
df = filter_crispr_by_model(df, df_map, minlines=10, line_group='OncotreeLineage')
df
```

# Select some tissues (semi supervised)
We start labelling genes for the `Kidney` tissue...

```python
from help.preprocess.imputer import imputer_knn
from sklearn.semi_supervised import LabelPropagation
Bagel_E = pd.read_csv(os.path.join(datapath, "curated_BAGEL_E.tsv"), index_col=0)
Bagel_NE = pd.read_csv(os.path.join(datapath, "curated_BAGEL_NE.tsv"), index_col=0)
label_prop_model = LabelPropagation(kernel='knn')
df1 = df[df.index.isin(Bagel_E.index)]
df1 = df1.assign(label=lambda x: 0)
df2 = df[df.index.isin(Bagel_NE.index)]
df2 = df2.assign(label=lambda x: 1)
df3 = df[~((df.index.isin(Bagel_E.index)) | (df.index.isin(Bagel_NE.index)))]
df3 = df3.assign(label=lambda x: -1)
df_m = pd.concat([df1,df2,df3], axis=0)
df_m = imputer_knn(df_m)
model = label_prop_model.fit(df_m.drop(columns=['label']), df_m['label'])
```

```python
df_m[df_m['label'] == -1].drop(columns=['label'])
labels = model.predict(df_m[df_m['label'] == -1].drop(columns=['label']))
df_m.loc[df_m.label==-1, 'label'] = labels
df_m['label'].value_counts()
```
