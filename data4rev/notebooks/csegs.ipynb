```python
import pandas as pd
import os
import numpy as np
from HELPpy.visualization.plot import svenn_intesect
path = '../data'
df = pd.read_csv(os.path.join(path, "HELP_PanCancer_threeClasses.csv"), index_col=0)
cEG = df[df['label']=='E'].index.values
ucs_EG1 = pd.read_csv(os.path.join(path, "ucsEG_Lung.txt"), header=None, index_col=0).index.values
df1 = pd.read_csv(os.path.join(path, "HELP_LungNeuro.csv"), index_col=0)
ucs_EG2 = df1[df1['label']=='E'].index.values
ucs_EG2 = np.setdiff1d(ucs_EG2, cEG)
df2 = pd.read_csv(os.path.join(path, "HELP_NSCLC.csv"), index_col=0)
ucs_EG3 = df2[df2['label']=='E'].index.values
ucs_EG3 = np.setdiff1d(ucs_EG3, cEG)
svenn_intesect([set(ucs_EG2),set(ucs_EG3), set(ucs_EG1)], labels=['HELP NE', 'HELP NSCLC', 'HELP Lung'], ylabel='ucsEGs EGs', figsize=(10,2))
```
